# 난이도 초급

```python
# 부모 클래스: 탈것
class Vehicle:
    def __init__(self, manufacture_date, speed=10):
        self.manufacture_date = manufacture_date
        self.speed = speed

    def move_forward(self):
        self.speed += 20
        print(f"{self.speed} 속도로 앞으로 이동합니다.")

    def move_backward(self):
        self.speed += 20
        print(f"{self.speed} 속도로 뒤로 이동합니다.")

    def move_right(self):
        self.speed += 20
        print(f"{self.speed} 속도로 오른쪽으로 이동합니다.")

    def move_left(self):
        self.speed += 20
        print(f"{self.speed} 속도로 왼쪽으로 이동합니다.")

# 자식 클래스: 할리
class Harley(Vehicle):
    def __init__(self, manufacture_date, speed=10, passengers=1):
        super().__init__(manufacture_date, speed)
        self.passengers = passengers

    def do_stunt(self):
        print("할리가 멋진 묘기를 합니다!")

# 자식 클래스: 스포츠카
class SportsCar(Vehicle):
    def __init__(self, manufacture_date, speed=10, passengers=2, horsepower=300):
        super().__init__(manufacture_date, speed)
        self.passengers = passengers
        self.horsepower = horsepower

    def overtake(self):
        self.speed *= 5
        print(str(self.speed) + " 속도로 스포츠카가 추월합니다!")

    def stop(self):
        self.speed = 0
        print(str(self.speed) + "스포츠카가 멈춥니다!")

    def honk(self):
        print("스포츠카가 빵빵거립니다!")

# 자식 클래스: 잠수함
class Submarine(Vehicle):
    def __init__(self, manufacture_date, speed=10, passengers=20):
        super().__init__(manufacture_date, speed)
        self.passengers = passengers

    def dive(self):
        print("잠수함이 잠수합니다!")

    def launch_missile(self):
        print("잠수함이 미사일을 발사합니다!")

# 예제 사용
harley = Harley("2022-01-01")
harley.move_forward()  # 30 속도로 앞으로 이동
harley.move_backward()  # 50 속도로 뒤로 이동
harley.do_stunt()

sports_car = SportsCar("2023-01-01")
sports_car.move_right()  # 30 속도로 오른쪽으로 이동
sports_car.overtake()
sports_car.honk()

submarine = Submarine("2020-01-01")
submarine.move_left()  # 30 속도로 왼쪽으로 이동
submarine.dive()
submarine.launch_missile()
```

## [소수 구별](https://freedeveloper.tistory.com/391)
프로그래머 X는 입력값을 숫자를 입력하거나 문자를 입력하려고 하는데,
- 만약 숫자를 입력하였으면 그것이 정수인지, 소수인지 구별하는 프로그램을 짜보도록 하고,
- 만약 문자를 입력하였으면 숫자가 아니므로 math error를 표시하게 하라.

### 정답

```python
import math

class NumberChecker:
    def __init__(self, inputs):
        self.inputs = inputs

    def is_prime_num(self, n):
        if n <= 1:
            return False  # 1 이하의 수는 소수가 아님
        for i in range(2, int(math.sqrt(n)) + 1):
            if n % i == 0:
                return False
        return True

    def check_number_type(self, value):
        try:
            num = int(value)
            if self.is_prime_num(num):
                return f"{num}은(는) 소수입니다."
            else:
                return f"{num}은(는) 소수가 아닙니다."
        except ValueError:
            return "math error"  # 변환에 실패하면 math error 표시

    def run_checks(self):
        for input_str in self.inputs:
            result = self.check_number_type(input_str)
            print(result)

# 테스트용 입력 리스트
inputs = ["-10", "0", "1", "10", "7", "5.5", "Hello World"]

# NumberChecker 클래스 인스턴스 생성 및 체크 실행
number_checker = NumberChecker(inputs)
number_checker.run_checks()
```

## [최대값, 최소값, 중간값 출력](https://codingdojang.com/scode/708#answer-filter-area)
- 예시 >> [8, 6, 2, 8, 5, 1, 2, 5, 4, 9]
  - 최소값: 1 / 최대값: 9 / 중간값: 5



### 정답

```python
import random
```

```python
class RandomList:
    def __init__(self, p_range=10):
        self.lst = [ random.randint(0,9) for _ in range(p_range)]

    def __call__(self):
        random.shuffle(self.lst)
        return self.lst
```

```python
class Game:
    def __init__(self):
        self.random_list = RandomList()
        self.median = 0
        self.minimum = 0
        self.maximum = 0

    def __process(self, lst):
        print(f'before list: {lst}')
        lst.sort() # 정렬
        print(f'after lst: {lst}')

        self.median = (len(lst) -1) / 2
        self.minimum = lst[0]
        self.maximum = lst[-1]

    def run(self, p_range=None):
        if p_range:
            self.random_list = RandomList(p_range)

        lst = self.random_list()
        self.__process(lst)

        print(f'최소값: {self.minimum} / 최대값: {self.maximum} / 중간값: {lst[int(self.median)]}')
```

```python
game = Game()

game.run()
```

```python
game.run(5)
```

## [Up&Down 숫자 맞추기](https://codingdojang.com/scode/711#answer-filter-area)
- 컴퓨터가 1 ~ 100 중 랜덤으로 숫자를 정합니다.
- 입력한 숫자보다 정답이 크면, "UP" 출력,
- 입력한 숫자보다 정답이 작으면, "DOWN" 출력,
- 정답을 맞추면, "정답" 출력

### 정답

```python
import random

class ParentGame:
    def __init__(self, p_range):
        self.p_range = p_range
        self.correct = 0
        self._mk_correct()

    def _mk_correct(self):
        self.correct = random.randint(1,self.p_range)
```

```python
class Game(ParentGame):
    def __init__(self, p_range=100):
        super().__init__(p_range)
        self.trial_num = 0
        self.answer = ''

    def __init(self):
        self.trial_num = 0
        self.answer = ''
        self._mk_correct()

    def __check(self):
        result = False

        if not self.answer.isdigit():
            print("숫자만 입력해주세요.")
            return result

        self.answer = int(self.answer)
        if  self.answer == self.correct:
            print(f"정답입니다. 입력한 횟수: {self.trial_num}")
            result = True
        elif self.answer > self.correct:
            print("DOWN")
        else:
            print("UP")

        return result

    def run(self):
        try:
            self.__init()
            print(f'correct: {self.correct}')

            while True:
                self.trial_num += 1
                self.answer = input("1~100 중 하나의 숫자를 입력해주세요 : ")
                if self.__check():
                    break
        except Exception as e:
            print(f'Error: {str(e)}')
```

```python
game = Game()

game.run()
```

```python
game.run()
```

## [방탈출 게임](https://codingdojang.com/scode/726#answer-filter-area)
- 비밀번호는 3자리로 한다.
- 비밀번호를 입력하게 한다.
- 비밀번호가 틀리면 처음부터 다시 입력하게 한다.
- 비밀번호가 맞으면 성공 메시지를 출력한다.

### 정답

```python
class MakePassword:
    def __init__(self):
        self.lst = [ str(i) for i in range(10) ]

    def _mk_password(self):
        _password = random.sample(self.lst, k=3)
        return "".join(_password)
```

```python
class ShowMSG:
    def _show_msg(self, answer, password):
      # 입력값 검증
      if len(answer) != 3:
          print("비밀번호 길이가 틀립니다.")
      elif not answer.isdigit():
          print("숫자만 입력해주세요.")

      # 정답 비교
      elif answer == password:
          print('성공')
      else:
          print('틀림')
```

```python
class Game(MakePassword, ShowMSG):
    def __call__(self):
        password = self._mk_password()
        print(f'password: {password}')
        answer = input()

        self._show_msg(answer, password)
```

```python
game = Game()

game()
```
